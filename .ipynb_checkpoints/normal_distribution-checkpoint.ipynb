This Jupyter notebook will explain the normal distribution in Mathematics & Statistics.  

The Normal distribution (also known as the Gaussian distribution) is an important tool that is used to show the probability distribution for randomly generated variables within datasets. 

```python
# importing NumPy to access numerical arrays
import numpy as np

# importing MatPlotLib to create histograms
import matplotlib.pyplot as plt

# importing Scipy Stats to utilise statistical functions
import scipy.stats as stats

# Creating a new, seeded random number generator
rng = np.random.default_rng(0)
# Generating 1,000,000 numbers on a standard normal distribution
samples = rng.standard_normal(1000000)

# Creating a figure
fig, ax = plt.subplots()
# Creating a histogram with the above random data
ax.hist(samples, bins = 100)
```

Analysing the information from the histogram above:  

Data points in a Normal Distribution plot are symmetrical about the mean (median/average) value from the random dataset, meaning that the data points in closer proximity to the mean value occur more frequently.  

Data points are distributed symmetrically as they tail off from the mean value, resulting in a bell-shaped curve. 

There exist a few aspects in life which make use of the Normal/Gaussian Distribution, examples included measuring height, blood pressure, marks on exams (as well as any measurement errors pertinent to each aspect).

The imported SciPy Stats module contains the function for normal distribution, stats.norm(mu, std):  

Here we plug in our values for mu/mean and std/standard deviation into the function above.  

The pdf() function stands for probability density function, this generates estimates for probability.

```python
# Setting values for mean and standard deviation
mu = 5
std = 10

# Plugging in values to normal distribution function
snd = stats.norm(mu, std)

# Generating 1,000 random values set between -100 & 100
x = np.linspace(-100, 100, 1000)

# Plotting the normal distribution for random values
plt.figure(figsize = (7.5, 7.5))
plt.plot(x, snd.pdf(x))
plt.xlim(-60, 60)
plt.title('Normal Distribution (Mean = 5, Standard Deviation = 10)', fontsize = '12')
plt.xlabel('Values of Random Variable (x)', fontsize = '12')
plt.ylabel('Probability', fontsize = '12')
plt.show()
```

We will now generate multiple plots with different values set for the mean and standard deviation (SD) to display the differences in how data points are distributed.

```python
# Generating 1,000 random values set between -100 & 100
x = np.linspace(-100, 100, 1000)
plt.figure(figsize = (7.5, 7.5))

# Plugging in different values for the mean and SD respectively
plt.plot(x, stats.norm(0, 1).pdf(x))
plt.plot(x, stats.norm(1, 0.75).pdf(x))
plt.plot(x, stats.norm(2, 1.5).pdf(x))

plt.xlim(-10, 10)
plt.title('Normal Distribution', fontsize = '12')
plt.xlabel('Values of Random Variable (x)', fontsize = '12')
plt.ylabel('Probability', fontsize = '12')
plt.legend(['mu 0 & std 1', 'mu 1 & std 0.75', 'mu 2 & std 1.5'])
plt.show()
```

Standard deviation (denoted by the Greek letter sigma ) is a property that is part of the Empirical Rule, where it describes the percentage of data points that are distributed further from the mean value, noted as units of standard deviation (SD) in the Normal Distribution curve.  

68% (2/3) of random data points occur within + or - 1 SD.  
95% (9.5/10) of random data points occur within + or - 2 SD.  
99.7% of random data points occur within + or - 3 SD.  

```python
# Generating 1,000 random values set between -100 & 100
x = np.linspace(-100, 100, 1000)
plt.figure(figsize = (7.5, 7.5))

# Plugging in different values for the mean and SD respectively
plt.plot(x, stats.norm(60, 10).pdf(x))
plt.plot(x, stats.norm(45, 5).pdf(x))
plt.plot(x, stats.norm(30, 5).pdf(x))

plt.xlim(-10, 90)
plt.title('Normal Distribution of Pizza Delivery Times', fontsize = '12')
plt.xlabel('Minutes', fontsize = '12')
plt.ylabel('Probability', fontsize = '12')
plt.legend(['mu 60 & std 10', 'mu 45 & std 5', 'mu 30 & std 5'])
plt.show()
```

The plot above details pizza delivery times and the standard deviations that can exist.  
On average, a pizza is delivered in 60 minutes (more or less), the standard deviation is the 'more or less' aspect, which is set to 10 minutes or 5 minutes.
